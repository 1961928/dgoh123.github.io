```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population"
table_class = "wikitable sortable"
```

```python
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
tables = soup.find('table',{'class':table_class})
```

```python
df = pd.read_html(str(tables))
```

```python
df[0]
```

```python
population = pd.concat(df, axis=1)
```

```python
population.columns = ['rank', 'Country', 'Population', '% of world', 'Date', 'Source', 'Notes']
```

```python
population
```

```python
population.drop(['rank', '% of world', 'Date', 'Source', 'Notes'], axis=1, inplace=True)
```

```python
population
```

```python
population.drop([0]).head(10).to_csv('population_top10.csv')
```
