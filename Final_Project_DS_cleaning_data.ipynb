import pandas to help me work with csv files in python

Read the files using list comprehension. I use the pandas read_csv function which tell pandas to parse any dates in the csv files into a pandas date time format.I also tell it to use the first column for the row indices and use a loop to do this for each of the files in fred_predictors.

```python
import pandas as pd

fred_predictors = ["CPI.csv", "GDP.csv", "mortgage_rate.csv", "real_income.csv", "rent_vac.csv", "house_supply.csv", "employment.csv", "cons_conf.csv"]
df = [pd.read_csv(i, parse_dates=True, index_col=0) for i in fred_predictors]
```

use the pd.concat function to combine the files into one dataframe.

```python
fred_merged = pd.concat(df, axis=1)
```

changing the column titles for easier interpretation of data.

```python
fred_merged.columns = ['CPI', "GDP", "Mortgage Rate", "Income", "Rental Vacancy Rate", "House Supply", "Employment Level", "Consumer Sentiment"]
```

```python
fred_merged
```

There is a lot of missing data as data is released on different dates. Some data such as the rental vancancy rate is released quarterly, inflation data is released monthly and data such as the real median income is released annually. I therefore assumed the value for the data such as the real median income was the same for the entire year until new data is released on the next year. I did this by using a forward fill ffill() which pulls values forward and fills in the missing values.

```python
fred_merged = fred_merged.ffill()
```

```python
fred_merged.tail(50)
```

I now read in the house price data. The first file has the median sales price of houses in the US for each week and the second files has the Zillow House Price Index released monthly.

```python
price_data = ["weekly_sales_price.csv", "monthly_home_values.csv"]
```

Read using pd.read_csv

```python
df = [pd.read_csv(i) for i in price_data]
```

iloc[0,5:] will select the first row of data and remove the first 5 columns. This will make the format of the data consistent with the FRED data with each row being a week and each column showing the sales prices of the house.

```python
df = [pd.DataFrame(i.iloc[0,5:]) for i in df]
```

The house price data is released weekly while the house values data is released monthly. We need to therefore add a column to both data frames and use it to merge them together using the common column.

The row indices showing the data is currently stored as a string so I convert it to a pandas datetime. i["Date"] = i.index.to_period("M") will remove the date part and will keep the month. Now there will be a new 'Date' column which has the year and month.

```python
for i in df:
  i.index = pd.to_datetime(i.index)
  i["Date"] = i.index.to_period("M")
```

 I use the 'Date' column to merge the data frames.

```python
price_data_merged = df[0].merge(df[1], on="Date")
```

Ensure the index of the data frame is the same as the index of the first data frame

```python
price_data_merged.index = df[0].index
```

Rename the columns to 'House Prices' and 'House Values' and remove the 'Date' column as it is now not needed.

```python
del price_data_merged["Date"]
price_data_merged.columns = ["House Prices", "House Values"]
```

Next step is to merge the price data with the FRED data

The Zillow data and the FRED data is released on different days. The Zillow data is released 2 days after the FRED data so I need to align the FRED data with the Zillow data. fred_merged.index = fred_merged.index + timedelta(days=2) will shift the FRED data forward by two days.

```python
from datetime import timedelta

fred_merged.index = fred_merged.index + timedelta(days=2)
```

```python
price_data_merged
```

Now that the dates are aligned, I can merge the FRED data with the Zillow data. The left_index=True, right_index=True is telling pandas to combine the data frames and use the index (the date) of both data frames to merge the data.

```python
price_data_merged = fred_merged.merge(price_data_merged, left_index=True, right_index=True)
```

```python
price_data_merged
```

Now that the data is cleaned, I download the file as a csv file.

```python
price_data_merged.to_csv('predictor_data2.csv')
```
